```python
# 3-ье Задание

import pandasql as ps

df_ratings = pd.read_csv(ratings)
df_ratings
```

```python
df_count_ratings = df_ratings.groupby(['userId'])['rating'].count().reset_index(name='count_ratings') # table to count ratings
df_count_ratings = df_count_ratings[(df_count_ratings['count_ratings'] > 100)]                        # and sort it
```

```python
def min_max(df):
    return df['timestamp'].max() - df['timestamp'].min()                                         # agg func

df_users_timestamp = df_ratings.groupby(['userId']).apply(min_max).reset_index(name='diff')      # table to get diff timestamp
```

```python
df_merge = pd.merge(left=df_count_ratings, right=df_users_timestamp, left_on='userId', right_on='userId')    # merge them
```

```python
from datetime import datetime
from datetime import date

timestamp = int(str(round(df_merge['diff'].mean())))               # get timestamp
str(datetime.utcfromtimestamp(timestamp) - datetime(1970, 1, 1))   # avg user live time
```
